# Optimizing Supply Chain in the Instant Noodles Business

## Business Problem

An FMCG company ventured into the instant noodles business two years ago. However, they have observed a significant mismatch between the demand and supply across their various warehouses in the country. In some regions, the demand for instant noodles is high, but the supply is insufficient, resulting in missed sales opportunities and customer dissatisfaction. In contrast, there are regions with low demand, but the supply exceeds the demand, leading to inventory cost losses.

The higher management has recognized that this imbalance in the supply chain is detrimental to the company's profitability. To address this issue, they want to optimize the supply quantity in every warehouse across the entire country. The objective is to align the supply with the local demand, minimize inventory costs, and enhance overall operational efficiency.

## Technical Approach

To tackle this business problem, we can follow a data-driven approach using various technical steps. Here's a high-level outline of how we can proceed:

1. **Data Collection**: Gather historical data on sales, demand, and supply for each warehouse. This data should be granular and cover at least the past two years to identify trends and patterns.

2. **Data Cleaning**: Preprocess the data to handle missing values, outliers, and inconsistencies. Ensure that the data is in a format suitable for analysis.

3. **Exploratory Data Analysis (EDA)**:
   - Explore the data to gain insights into demand and supply trends.
   - Identify regions with high demand and low supply, as well as regions with low demand and excess supply.

4. **Demand Forecasting**:
   - Use time series analysis or forecasting models to predict future demand for each warehouse.
   - Validate the accuracy of the forecasts to ensure reliability.

5. **Supply Chain Optimization**:
   - Utilize optimization algorithms to determine the optimal supply quantity for each warehouse.
   - Consider factors such as demand forecasts, transportation costs, storage costs, and budget constraints.
   
6. **Feature Engineering:** Create or transform features that can improve the predictive power of the model.


7. **Model Building:** Train and evaluate machine learning models for predicting the probability of default. Common models include logistic regression, decision trees, random forests, and gradient boosting.


8. **Model Evaluation:** Use the metrics to assess the model's performance.


9. **Model Deployment:** If the model performs well, consider deploying it for real-time credit risk assessment.


10. **Documentation and Reporting:** Document the project's findings, including model performance, insights from EDA, and recommendations for risk management.

By following this technical approach, the FMCG company can optimize its supply chain, reduce inventory costs, and improve customer satisfaction by ensuring that the right amount of instant noodles is available in each warehouse to meet local demand.

In the subsequent sections of this Jupyter Notebook, we will delve into each of these steps in detail, applying data analysis, forecasting, and optimization techniques to address the business problem effectively.


# Feature Descriptions

In this section, we'll provide detailed explanations for each of the features in the dataset, which are related to the FMCG company's supply chain and warehouse operations.

1. **Warehouse_id**: Product warehouse_id
<br/><br/>

2. **WH_Manager_ID**: Identifier for the warehouse manager, responsible for the day-to-day operations and decision-making within the warehouse.
<br/><br/>

3. **Location_type**: Describes the type of location where the warehouse is situated (e.g., urban, rural).
<br/><br/>

4. **WH_capacity_size**: Indicates the capacity or size of the warehouse, which affects its storage capabilities.
<br/><br/>

5. **Zone**: Zone of the warehouse
<br/><br/>

6. **WH_regional_zone**: Refers to the regional zone where the warehouse is located, which can have implications for supply chain planning and distribution.
<br/><br/>

7. **num_refill_req_l3m**: The number of times refills have been requested in the last three months, which is a measure of inventory management.
<br/><br/>

8. **transport_issue_l1y**: Indicates whether the warehouse has experienced transport-related issues in the last year, which could impact the supply chain.
<br/><br/>

9. **Competitor_in_mkt**: Specifies whether competitors are present in the market, which can influence demand and supply dynamics.
<br/><br/>

10. **retail_shop_num**: The number of retail shops serviced by the warehouse, influencing the demand for products.
<br/><br/>

11. **wh_owner_type**: Describes the type of ownership for the warehouse (e.g., company-owned, rent).
<br/><br/>

12. **distributor_num**: The number of distributors associated with the warehouse, which affects the distribution network.
<br/><br/>

13. **flood_impacted**: Indicates whether the warehouse has been impacted by floods, a factor that could affect operations and require flood-proofing.
<br/><br/>

14. **Flood_proof**: Specifies whether the warehouse is flood-proof, which is crucial for maintaining inventory during flood events.
<br/><br/>

15. **electric_supply**: Warehouse have electric back up like generator, so they can run the warehouse in load shedding
<br/><br/>

16. **dist_from_hub**: The distance of the warehouse from a central hub or distribution center, influencing transportation costs and logistics.
<br/><br/>

17. **workers_num**: The number of workers employed at the warehouse, which impacts operational efficiency.
<br/><br/>

18. **wh_est_year**: The year in which the warehouse was established, which can provide insights into its age and infrastructure.
<br/><br/>

19. **storage_issue_reported_l3m**: Warehouse reported storage issue to corporate office in last 3 months. Like rat, fungus because of moisture etc.
<br/><br/>

20. **temp_reg_mach**: The presence of temperature-regulating machinery, which is essential for storing certain products.
<br/><br/>

21. **approved_wh_govt_certificate**: Indicates whether the warehouse has a government-approved certificate, which can affect its legality and reliability.
<br/><br/>

22. **wh_breakdown_l3m**: Number of time warehouse face a breakdown in last 3 months. Like strike from worker, flood, or electrical failure
<br/><br/>

23. **govt_check_l3m**: Number of time government Officers have been visited the warehouse to check the quality and expire of stored food in last 3 months
<br/><br/>

24. **product_wg_ton (Target_column)**: Product has been shipped in last 3 months. Weight is in tons
<br/><br/>

These features collectively provide valuable information about each warehouse, its operations, and its surrounding environment. Analyzing these features can help the company make data-driven decisions to optimize the supply chain and enhance warehouse performance.

```python
def missing_values(df):
    missing_number = df.isnull().sum().sort_values(ascending=False)
    missing_percent = (df.isnull().sum()/df.shape[0]).sort_values(ascending=False)
    missing_values = pd.concat([missing_number, missing_percent], axis=1, keys=['Missing_Number', 'Missing_Percent'])
    return missing_values[missing_values['Missing_Number']>0]
```

```python
def first_looking(df):
    print(colored("Shape:", attrs=['bold']), df.shape,'\n', 
          colored('*'*100, 'red', attrs=['bold']),
          colored("\nInfo:\n", attrs=['bold']), sep='')
    print(df.info(), '\n', 
          colored('*'*100, 'red', attrs=['bold']), sep='')
    print(colored("Number of Uniques:\n", attrs=['bold']), df.nunique(),'\n',
          colored('*'*100, 'red', attrs=['bold']), sep='')
    print(colored("Missing Values:\n", attrs=['bold']), missing_values(df),'\n', 
          colored('*'*100, 'red', attrs=['bold']), sep='')
    print(colored("All Columns:", attrs=['bold']), list(df.columns),'\n', 
          colored('*'*100, 'red', attrs=['bold']), sep='')

    df.columns= df.columns.str.lower().str.replace('&', '_').str.replace(' ', '_')
    print(colored("Columns after rename:", attrs=['bold']), list(df.columns),'\n',
          colored('*'*100, 'red', attrs=['bold']), sep='')
    print(colored("Descriptive Statistics \n", attrs=['bold']), df.describe().round(2),'\n',
          colored('*'*100, 'red', attrs=['bold']), sep='') # Gives a statstical breakdown of the data.
    print(colored("Descriptive Statistics (Categorical Columns) \n", attrs=['bold']), df.describe(include=object).T,'\n',
          colored('*'*100, 'red', attrs=['bold']), sep='') # Gives a statstical breakdown of the data.
```

```python
# To view summary information about the columns

def column_look(df,col):
    print("column name    : ", col)
    print("per_of_nulls   : ", "%", round(df[col].isnull().sum()/df.shape[0]*100, 2))
    print("num_of_nulls   : ", df[col].isnull().sum())
    print("num_of_uniques : ", df[col].nunique())
    print("Value counts   : \n", df[col].value_counts(dropna = False)) 
    print(colored("-"*100, 'red',attrs=['bold']), sep="")
```

```python
def multicolinearity_control(df, threshold=0.75):
    features = df.corr(numeric_only=True).columns
    collinear_pairs = []

    for i in range(len(features)):
        for j in range(i + 1, len(features)):
            corr = df.corr(numeric_only=True).iloc[i, j]
            if abs(corr) > threshold:
                collinear_pairs.append((features[i], features[j]))
                print(colored(f"Multicolinearity alert between: {features[i]} - {features[j]}, Correlation: {corr}","red", attrs=['bold']),'\n',
                                  colored('*'*100, 'red', attrs=['bold']), sep='')

    if not collinear_pairs:
        print(colored(f"No Multicolinearity detected.","red", attrs=['bold']),)
```

```python
def duplicate_values(df):
    print(colored("Duplicate check...", attrs=['bold']), sep='')
    duplicate_values = df.duplicated(subset=None, keep='first').sum()
    if duplicate_values > 0:
        df.drop_duplicates(keep='first', inplace=True)
        print(duplicate_values, colored(" Duplicates were dropped!"),'\n',
              colored('*'*100, 'red', attrs=['bold']), sep='')
    else:
        print(colored("There are no duplicates"),'\n',
              colored('*'*100, 'red', attrs=['bold']), sep='')
```

```python
def drop_columns(df, drop_columns):
    if drop_columns !=[]:
        df.drop(drop_columns, axis=1, inplace=True)
        print(drop_columns, 'were dropped')
    else:
        print(colored('We will now check the missing values and if necessary, the realted columns will be dropped!', attrs=['bold']),'\n',
              colored('*'*100, 'red', attrs=['bold']), sep='')
```

```python
def drop_null(df, limit):
    print('Shape:', df.shape)
    for i in df.isnull().sum().index:
        if (df.isnull().sum()[i]/df.shape[0]*100)>limit:
            print(df.isnull().sum()[i], 'percent of', i ,'null and were dropped')
            df.drop(i, axis=1, inplace=True)
            print('new shape:', df.shape)       
    print(colored("New shape after missing value control:"),'\n', df.shape)
```
